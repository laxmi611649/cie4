```python
import pandas as pd
df=pd.read_csv('/content/train.csv')
df.head()
df=df.drop(['Name','Sex','Ticket','Cabin','Embarked','Fare'],axis=1)
df=df.dropna()
df.head()
```

```python
target=df['Pclass']
train=df.drop('Pclass',axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(train,target,test_size=0.20)
```

```python
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
model1 = LinearRegression()
model2 = XGBRegressor()
model3 = RandomForestRegressor()
```

```python
model1.fit(x_train,y_train)
model2.fit(x_train,y_train)
model3.fit(x_train,y_train)
```

```python
pred1=model1.predict(x_test)
pred2=model2.predict(x_test)
pred3=model3.predict(x_test)
```

```python
final_pred=(pred1+pred2+pred3)/3.0
final_pred
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,final_pred))
```
